使用正则表达式高效处理字符串

在Python中需要通过正则表达式对字符串进行匹配的时候，可以使用一个模块，名字为re

```python
# coding = utf-8
import re
```

## 匹配出以xxx开头的字符串:match()

语法初步了解

```python
# 使用match方法进行匹配操作：re.match（）能够匹配出以xxx开头的字符串
# re.match（正则表达式，要匹配的字符串）

# 匹配以python开头的语句


res = re.match("python", "pythonlearniingday21")
print(res)
#如果上一步匹配到数据的话，可以使用group方法来提取数据
print(res.group())

res = re.match("python","mypythonlearniingday21")
print(res)# 未匹配到数据
# 注意None执行group()会报错
print(res.group())# 报错
```

### 匹配单个字符

```python
#coding=utf-8
import re

# 1、.匹配任意1个字符（除了\n）
res = re.match(".", "M")
print(res.group())

res = re.match(".", "python")
print(res.group())

res = re.match("t.o", "too")
print(res.group())

res = re.match("t.o", "two")
print(res.group())

res = re.match("t.o", "ppptoo")
print(res)# 注意None执行group()会报错

# 要匹配.需加转义字符\
res = re.match("\.", ".pynb_checkpoints/")
print(res.group())
```

```python
# 2、匹配[]中列举的字符

## 2.1 大小写匹配
res = re.match("h", "hello world")
print(res.group())

res = re.match("H", "Hello world")
print(res.group())

# 大小写都可以
res = re.match("[hH]", "hello world")
print(res.group())

res = re.match("[hH]", "Hello world")
print(res.group())

res = re.match("[hH]ello w", "Hello world")
print(res.group())

res = re.match("[hH]ello [wW]orld", "Hello world")
print(res.group())


# 2.2 匹配大小写字母：A-Z使用-表示在ASCII码表上是连续的
res = re.match("[A-Z]", "Python Learning")
print(res.group())

res = re.match("[a-z]", "python Learning")
print(res.group())

# 大小写都可以
res = re.match("[a-zA-Z]", "Python Learning")
print(res.group())

## 2.3 匹配0-9
### 方法一
res = re.match("[0123456789]Hello w","6Hello world")
print(res.group())
### 方法二
res = re.match("[0-9]Hello w", "6Hello world")
print(res.group())

# 2.4 匹配0-9（除6）
res = re.match("[0-57-9]Hello w", "6Hello world")
print(res)
```

```python
# 匹配多个字符

# 匹配开头两个字符：第一个为大写，第二个为小写
res = re.match("[A-Z][a-z]", "Python learning")
print(res.group())

res = re.match("[a-z][A-Z]", "pYthon leaning")
print(res.group())
```

```python
# 3、\d匹配数字，即0-9dicimal
res = re.match("嫦娥1号", "嫦娥1号发射成功")
print(res.group())
res = re.match("嫦娥2号", "嫦娥2号发射成功")
print(res.group())

res = re.match("嫦娥\d号", "嫦娥3号发射成功")
print(res.group())
```

```python
# \w匹配单词字符：a-z，A-Z，0-9，_(下划线)，汉字
res = re.match("[\w]", "name1  python")
print(res.group())

res = re.match("[\w]", "名字name  python")
print(res.group())

res = re.match("[\w]", "__name  python")
print(res.group())

res = re.match("[\w]", "00name  python")
print(res.group())

res = re.match("[\w]", ".....name  python")# 标点符号非单词
print(res)
```

```python
# \w匹配非单词字符
res = re.match("[\W]", ".....name  python")# 标点符号非单词
print(res.group())

res = re.match("[\W]", "name  python")# 标点符号非单词
print(res)
```

### 匹配多个字符

需求：匹配出，一个字符串第一个字母为大小字符，后面都是小写字母并且这些小写字母可有可无


```python
# coding=utf-8
import re
# 1、 *前的字符出现0次/无限次（可有可无）
print()
res = re.match("[A-Z][a-z]", "Python learning")# 只能匹配开头两个字符：第一个为大写，第二个为小写
print(res.group())

res = re.match("[A-Z][a-z]*", "Python learning")# * 前的小写字母[a-z]可有0-多个，遇到空格则结束
print(res.group())

res = re.match("[A-Z][a-z]*", "Pythonlearning hello")
print(res.group())

res = re.match("[A-Z][a-z]*", "PYTHON LEARNING")# 小写字母可无
print(res.group())
```

需求：匹配出，变量名是否有效

```python
# 2、 +前的字符出现至少出现1次

res = re.match("[a-zA-Z]+","name python")# [a-zA-Z]+必须以大写或小写字母开头--大小写字母必须出现至少一次
print(res.group())

res = re.match("[a-zA-Z]+","_name python")# 大小写字母必须出现至少一次
print(res)
# [a-zA-Z]+必须以大写或小写字母开头(直到遇到非字母)且紧接着的非字母只能出现非单词字符
res =re.match("[a-zA-Z]+[\W]","name,,python" )
print(res.group())

res =re.match("[a-zA-Z]+[\W]","name000python" )
print(res)


# [a-zA-Z]+必须以大写或小写字母开头(直到遇到非字母)且紧接着的非字母只能出现单词字符
res =re.match("[a-zA-Z]+[\w]*","name,,python" )
print(res.group())

res =re.match("[a-zA-Z]+[\w]*","name_0_0_0。。。python" )
print(res.group())

# 必须以大写或小写字母开头或下划线_(直到遇到非字母)且紧接着的非字母只能出现单词字符
res = re.match("[a-zA-Z_]+[\w]*", "_name")
print(res.group())
```

```python
names = ["name1","_name","2_name","_name_"]
for name in names:
    res = re.match("[a-zA-Z_]+[\w]*", name)
    if res:
        print("变量名%s:符号要求" % res.group())
    else:
        print("变量名%s:非法" % name)
```

需求：匹配出0-99之间的数字

```python
# 匹配出0-99之间的数字:如果是个位数，开头不能是0
## 方法一
res = re.match("[1-9]?[0-9]", "77")# 如果是个位数，开头不能是0，只能是1-9，后面可以是任意数字
print(res.group())

res = re.match("[1-9]?[0-9]", "0") 
print(res.group())

## 方法二
res = re.match("[1-9]?\d","199")# 第三位9不会匹配
print(res.group())

# 个位数，开头是0，怎么处理--使用$解决
res = re.match("[1-9]?[0-9]", "09")
print(res.group())# 开头是0， 结果错误

# 使用$ 使错误匹配报错，防止错误匹配的输出
res = re.match("[1-9]?\d$", "08")
print(res)
```

题目1：匹配出163的邮箱地址，且@符号之前有4到20位，例如hello@163.com


```python
# 使用之前匹配开头的方法
res = re.match("[\w]{4,20}@163\.com","xiaowang@163.com" )# 在.前面加\对其进行转义
print(res.group())

# 结尾无法匹配，会出现错误匹配
res = re.match("[\w]{4,20}@163\.com","xiaowang@163.comqqqq" )# 在.前面加\对其进行转义
print(res.group())

# 使用$限定以m结尾--(注意match本身是从开头匹配的)
res = re.match("[\w]{4,20}@163\.com$","xiaowang@163.comqqqq" )# 在.前面加\对其进行转义
print(res)

res = re.match("[\w]{4,20}@163\.com$","xiaowang@163.com" )# 在.前面加\对其进行转义
print(res.group())
```

### 匹配分组

匹配0-100之间的数字

```python
# 之前的匹配0-99方法无法匹配0开头的个位数情况,出现错误匹配
res = re.match("[1-9]?\d", "08")
print(res.group())# 0
# 使用$ 使错误匹配报错，防止错误匹配的输出
res = re.match("[1-9]?\d$", "08")
print(res)

# 现在匹配0-100，100是三位数
res = re.match("[1-9]?\d$|100", "100")# |类似于：或，匹配左边或右边成功即可
print(res.group())
```

需求：匹配出163、126、qq邮箱

```python
# 只匹配163
res = re.match("[\w]{4,20}@163\.com$","xiaowang@163.com" )# 在.前面加\对其进行转义
print(res.group())

# 只匹配126
res = re.match("[\w]{4,20}@126\.com$","xiaowang@126.com" )# 在.前面加\对其进行转义
print(res.group())

# 只匹配qq
res = re.match("[\w]{4,20}@qq\.com$","xiaowang@qq.com" )# 在.前面加\对其进行转义
print(res.group())

# 匹配163/126/qq--加括号进行分组
res = re.match("\w{4,20}@(163|126|qq)\.com", "xiaowang@qq.com")
print(res.group())
```

需求：提取区号和手机号码

```python
# 使用分组分别提取
res = re.match("([^-]+)-(\d+)", "010-1000001")
print(res.group())#区号和手机号码
print(res.group(1))#区号
print(res.group(2))#手机号码
```

```python
# 需求：匹配出"<html>hh</html>"--前面的<>和后面的<>内容保持一致


res = re.match("<[a-zA-Z]*>\w*</[a-zA-Z]*>", "<html>hh</html>")
print(res.group())

# 得到错误的匹配结果-->前后<>的内容不一致
res = re.match("<[a-zA-Z]*>\w*</[a-zA-Z]*>","<html>hh</htmllalala>" )
print(res.group())

# 使用\引用分组中匹配到的数据解决
res = re.match(r"<([a-zA-Z]*)>\w*</\1>","<html>hh</htmllalala>")# 第二个分组中只引用第一个分组的内容
print(res)
```

```python
# 需求：匹配出<htm1><h1>www.cskaoyan.com</h1></html>

res = re.match(r"<(\w*)><(\w*)>.*</\2></\1>", "<html><h1>www. cskaoyan. com</h1></html>")
print(res.group())

res = re.match(r"<(\w*)><(\w*)>.*</\2></\1>","<html><h1>www. cskaoyan. com</h2></html>")
print(res)
```

为避免分组删除后修改上述代码麻烦，使用分组别名

```python
res = re.match(r"<(?P<g1>\w*)><(?P<g2>\w*)>.*</(?P=g2)></(?P=g1)>", "<html><h1>www. cskaoyan. com</h1></html>")
print(res.group())
```

## findall()

```python
res = re.findall(r"\d+","阅读次数为900， 点赞次数为300")
print(res)# 返回的是列表
```

```python
# 当存在分组时只提取分组内容

s = "hello world，now is 2020/7/20 18:48，现在是2020/7/20 18:48。"
# 提取上述字符串的所有日期
pat = r"\d{4}/[01]?[0-9]/[1-3]?[0-9]\s(0[0-9]|1[0-9]|2[0-4]):[0-5][0-9]"
res = re.findall(pat, s)
print(res)# 错误结果：只提取了括号内的小时

# 要在括号内加？:
pat = r"\d{4}/[01]?[0-9]/[1-3]?[0-9]\s(?:0[0-9]|1[0-9]|2[0-4]):[0-5][0-9]"
res = re.findall(pat, s)
print(res)# 正确结果
```

## sub()

替换：sub(pat, repl, str)

句子 str 中找到匹配正则表达式模式的所有子字符串，用另一个字符串

repl 进行替换。如果没有找到匹配模式的串，则返回未被修改的句子 

str，其中 repl 既可以是字符串也可以是一个函数。

```python
s = "hello world，now is 2020/7/2018：48，现在是2020年7月20日18时48分。"

res1 = re.sub(r"年|月", r"/", s) # 将年、月替换成/
print(res1)

res2 = re.sub(r"日",r" ", res1)
print(res2)

res3 = re.sub(r"时", r":",res2)
print(res3)

res4 = re.sub(r"分", r"", res3)
print(res4)
```

```python
# sub是全部替换
res = re.sub(r"\d+", "998", "阅读次数为900， 点赞次数为300")
print(res)
```

```python
# 替换内容repl 既可以是字符串也可以是一个函数
def add_func(tmp):
    """
    将sub中提取的数字+1
    tmp,并非sub中直接匹配\d+的数字，需要进行提取
    tmp -- 是match（）/findall返回的对象--已经完成匹配的部分   
    """
    str_tmp = tmp.group()# 通过group()提取出来的是字符串
    print(str_tmp,type(str_tmp))
    num = int(str_tmp) + 1
    print(num,type(num))
    return str(num)

res = re.sub(r"\d+", add_func,"阅读次数为900, 点赞次数为300")
print(res)
```

## compile(pat)：将某个模式编译成对象，供之后使用

把正则表达式的模式转化成正则表达式对象，供其他函数如match 和 

search 使用。对象创建出来可以循环使用，如果某种模式要重复使用

话，用“先 compile 再 findall”的方式更加高效。

```python

email = '''Shengyuan Personal: quantsteven@gmail.com
Shengyuan Work: shengyuan@octagon-advisors.com
Shengyuan School: g0700508@nus.edu.sg
Obama: barack.obama@whitehouse.gov'''
print(email)

# 之前的匹配方法
res = re.match("[\w.-]+@[\w.-]+","quantsteven@gmail.com" )
print(res.group())
```

```python
# 使用compile的方式
# 创建电邮的模式 r'[\w.-]+@[\w.-]+'，用 compile 先创建 RE 对象，供之后使用。
pat = r"[\w.-]+@[\w.-]+"
com_obj = re.compile(pat)
print(com_obj)
#在对象 obj 上分别使用 match, search, findall, findieter 等方法
print(com_obj.match(email),"\n")
print(com_obj.search(email), "\n")
print(com_obj.findall(email), "\n")
print(com_obj.sub("---@---.---", email), "\n")
```

## 非贪婪操作符？

```python
s = "This is a number 234-235-22-423"
# 贪婪(越多越好)：.+是贪婪的，一直取最长的满足字符，/d只要还能匹配就可以了，所以只留了一个数字给第一个\d
res = re.match(".+(\d+-\d+-\d+-\d+)", s)
print(res.group())
print(res.group(1))
# 改为非贪婪(越少越好)：在*+？等字符后面加？将这些字符改为非贪婪
res = re.match(".+?(\d+-\d+-\d+-\d+)", s)
print(res.group())
print(res.group(1))
```

```python
res = re.match(r"aa(\d+)", "aa2343ddd")
print(res.group())
print(res.group(1))

res = re.match(r"aa(\d+?)", "aa2343ddd")
print(res.group())
print(res.group(1))
```

```python
res = re.match(r"aa(\d+)ddd", "aa2343ddd")
print(res.group())
print(res.group(1))

res = re.match(r"aa(\d+?)ddd", "aa2343ddd")
print(res.group())
print(res.group(1))
```

```python
text = """<img data-original="https://rpic.douyucdn.cn/appCovers/2016/11/13/1213973_201611131917_small.jpg" src="https://rpic.douyucdn.cn/appCovers/2016/11/13/1213973_201611131917_small.jpg" style="display: inline;">"""
# 只提取一个
res1 = re.search(r"https://.*?\.jpg", text)
print(res1.group())
# 提取全部
res2 = re.findall(r"https://.*?\.jpg", text)
print(res2)

# 不要最后的文件名，只需要主域名
res = re.search(r"https://.*\/", res1.group())
print(res.group())

res = re.search(r"https://.*?\/",res1.group())
print(res.group())
```

## r的作用

处理转义字符\的问题

```python
mm = "c:\\a\\b\\c"
print(mm)
mm_r = r"c:\\a\\b\\c"
print(mm_r)

# 匹配字符1个"\"需要4个反斜杠
res = re.match("c:\\\\",mm)
print(res.group())
# 正则使用r后匹配字符1个"\"只需要2个反斜杠
res = re.match(r"c:\\",mm)
print(res.group())

res = re.match(r"c:\\", mm_r)
print(res.group())
```

## 邮件匹配

匹配更多邮箱:

符合条件

    wan_long@vobile.cn
    07281070@bjtu.edu.cn
    xiaowang@163.com
    xiaowang@qq.com
    
不符合条件：

    .com.xiaowang@qq.com
    xiaowang@163.comqqqq
    
特定点:前面有数字/字符，中间有@，后面.xxx，然后空格结尾

前面：
    包含数字/字符/特殊符号：[A-Za-z0-9],[.-_]
    
    必须数字/字母开头[A-Za-z0-9]+
    不能以特殊字符开头：[A-Za-z0-9]+[.-_]
    必须@前面紧接数字/字符，不能接特殊字符：[A-Za-z0-9]+@
结果：[A-Za-z0-9]+[.-_])*[A-Za-z0-9]+@


@后面的域名：
    
    @后面必须紧接字母/数字：@[A-Za-z0-9]+
    .域名中，域名或全大写或全小写(\.[A-Z|a-z])，域名包含2个及其以上字母(\.[A-Z|a-z]{2，})
    
    
    
    

```python
model1 = re.compile(r'([A-Za-z0-9]+[.-_])*[A-Za-z0-9]+@[A-Za-z0-9-]+(\.[A-Z|a-z]{2,})+')
email_list = ["wan_long@vobile.cn","skyvense@online.sh.cn",".com.xiaowang@qq.com","07281070@bjtu.edu.cn","<dataguru@phx1-ss-2-lb.cnet.com>"]
for email in email_list:
    res = re.fullmatch(model1,email)
    if res:
        print("%s:符合要求" % res.group())
    else:
        print("%s:不符合要求" % email)
```

对于：

    "<dataguru@phx1-ss-2-lb.cnet.com>"也应该能提取出对应的邮箱

```python
# 使用match实现
model1 = re.compile(r'(?:[A-Za-z0-9]+[.-_])*[A-Za-z0-9]+@[A-Za-z0-9-]+(?:\.[A-Z|a-z]{2,})+')

email_list = ["b1063710427@ssmail.hit.edu.cn","dataguru@phx1-ss-2-lb.cnet.com",".com.xiaowang@qq.com","07281070@bjtu.edu.cn","<dataguru@phx1-ss-2-lb.cnet.com>"]
for email in email_list:
    res = re.findall(model1,email)
    if res:
        print("%s:符合要求" % res)
    else:
        print("%s:不符合要求" % email)
```

## 提取IP地址



32位ip地址：202.206.240.8
    
一个数字：

    每一个.前面的数字范围0-255，不以0开头:
    所以25[0-5]|2[0-4]\d|1\d\d|[1-9]?
    
    

```python
model_ip = re.compile(r'(?:(?:[1-9]?\d|1\d\d|2[0-4]\d|25[0-5])\.){3}(?:[1-9]?\d|1\d\d|2[0-4]\d|25[0-5])')
ip_list = ["222.178.10.26_accepted_message./Remote_host_said:_250_OK:has_queued/","Connected_to_202.96.74.68_but_sender_was_rejected./Remote_host_said:_554_Connection_refused(mx)._MAIL_FROM_[dataguru@phx1-ss-2-lb.cnet.com]_mismatches_client_IP_[61.55.167.172]./"]
for ip in ip_list:
    res = re.findall(model_ip, ip)
    if res:
        print("%s 符合要求" % res)
    else:
        print("%s 不符合要求" % ip)
```

```python
import re
# 包含开始信息的匹配模式:开始时有remote或local两种
pattern_start = re.compile(r".* starting delivery (.*): msg .* to (remote|local) (.*)")
pattern_response = re.compile(r".* delivery (.*): (.*): (.+?)/.*")
pattern_email = re.compile(r"(?:[A-Za-z0-9]+[.-_])*[A-Za-z0-9]+@[A-Za-z0-9-]+(?:\.[A-Z|a-z]{2,})+")
pattern_IP = re.compile(r'(?:(?:[1-9]?\d|1\d\d|2[0-4]\d|25[0-5])\.){3}(?:[1-9]?\d|1\d\d|2[0-4]\d|25[0-5])')

# 提取的信息记录：{'id':{'Email':具体邮箱号, "kind":具体种类}}
records_dict = {}  # eg：{' 1820': {'Email': 'sunkang@189.cn', 'kind': 'remote'}}

def search_line(line):
    """
    :param line: log文件的每一行
    每一行可能为开始传送/回复两种不同信息
    
    """
    # 每一行可能为开始传送/回复两种不同信息
    msg_start = pattern_start.findall(line)
    msg_response = pattern_response.findall(line)
    
    # 根据上述信息提取进行统计
    if msg_start:
        # 如果能提取到start信息
        post_id = msg_start[0][0]
        email = pattern_email.findall(msg_start[0][2])  # 返回的是一个列表,注意可能提取不到email
#         print(email)
        # 根据id作为字典的key--注意字典覆盖的情况
        if post_id not in records_dict:
            records_dict[post_id] = {"Email":email[0] if email else None}
        else:
            records_dict[post_id]["Email"] = email[0] if email else None
    
    if msg_response:
        # 如果能提取到response信息
        post_id = msg_response[0][0]
        status = msg_response[0][1]
        ip = pattern_IP.findall(msg_response[0][2])  # 返回的是列表,注意可能提取不到ip
        if post_id not in records_dict:
            records_dict[post_id] = {"status":status,"ip":ip[0] if ip else None}
        else:
            records_dict[post_id]["status"] = status
            records_dict[post_id]["ip"] = ip

        
        
def count_fail():
    # 统计失败信息
    fail_emails = {}  # 出现错误的邮箱记录
    for post_id,status_ip_email in records_dict.items():
#         print(post_id)
#         print(status_ip_email)
        if "status" in status_ip_email:
            if status_ip_email["status"] != "success" and "Email" in status_ip_email:
                fail_email = re.findall(r"@[A-Za-z0-9-]+(?:\.[A-Z|a-z]{2,})+", status_ip_email["Email"])[0]  # 出现错误的邮箱后缀
#                 print(fail_email)
                if fail_emails.get(fail_email):
                    fail_emails[fail_email] += 1
                else:
                    fail_emails[fail_email] = 1
    
    sorted_mails = sorted(fail_emails.items(), key=lambda d: d[1], reverse=True)
                    
     # 找出出现错误最多的那个
#     print(fail_emails)
#     print(sorted_mails)
    return sorted_mails[0][0]
    
```

```python
if __name__ == "__main__":
    # 带分析文件路径
    file = "./data/maillog.txt"
    try:
        with open(file, mode="r", encoding="utf-8") as f:
            for line in f:  # 逐行处理文件
                search_line(line)  # 保存每一行读取的信息
    except Exception as e:
        print(e)
        
    most_freq_emails = count_fail()
    print(most_freq_emails)
                
```
