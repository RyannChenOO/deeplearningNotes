# 打开文件



**使用文件，都得先打开文件**

`open(file=,mode=,encoding=)`:

    file--传入文件名字符串
    mode = 打开的模式

传入的文件名是一个包含路径的字符串:

    只给一个文件名时，默认当前编写程序时所在的文件中（当前目录下）查找文件
    可以使用包含相对路径的文件名到当前文件的子文件中查找文件--
    使用绝对路径到任何文件中查找文件
    


    要打开file_learning.txt需要使用相对路径：/data/file_learning.txt
    如果直接使用:file_learing.txt会在python_basic中找，找不到
    
    在Linux和OS X中，路径分割符是斜杠/
    在Windows系统中，在文件路径中使用反斜杠（ \ ）而不是斜杠（ / ），有时候斜杠（ / ）也可以--Windows系统有时能够正确地解读文件路径中的斜杠。如果你使用的是Windows系统，且结果不符合预期，请确保在文件路径中使用的是反斜杠
    
open() as 文件对象名：

函数 open()返回一个表示文件的对象, 这个对象存储在as后面声明的变量中

`with`关键字：

    使用with关键字，Python自会在合适的时候自动将其关闭。不需要自己调用close()
    如果不使用with，使用open（）打开文件后，需要自己close()关闭
    使用关键字 with 时， open() 返回的文件对象只在 with 代码块内可用

```python
with open(file="data/file_learning.txt", mode="r", encoding="utf-8") as file_object:# 默认以只读方式打开
    content = file_object.read()
    print(content)
```

## 按行读取

### 直接对文件对象使用for循环

```python
with open("data/file_learning.txt", encoding="utf-8") as f:
    i = 0
    for line in f:# 每行的末尾都有一个看不见的换行符
        print(i, line) # print也有一个换行符，所以最终每一行有两个换行符
        i +=1
```

```python
with open(file="data/file_learning.txt", encoding="utf-8", mode="r") as f:
    for line in f:# 每一行因为原始输入会有\n
        print(line, end="") # print后没有换行符，因此最后共有一个换行符 
        
```

每行的末尾都有一个看不见的换行符

每行的末尾都有一个看不见的换行符,使用rstrip()去掉

```python
with open("data/file_learning.txt", mode="r", encoding="utf-8") as f:
    i = 0
    for line in f:#文件对象是一个可迭代对象,按行处理
        print(i,line.rstrip()) # 注意每行的末尾都有一个看不见的换行符
        i += 1
```

### readline()

```python
with open(file="data/file_learning.txt", encoding="utf-8", mode="r") as f:
    line = f.readline()  # 只读取了第一行

print(line)
```

```python
with open(file="data/file_learning.txt", mode="r", encoding="utf-8") as f:
    line1 = f.readline()# 读取第一行，注意也会把换行符读进去
    line2 = f.readline()
    
print(line1)  # 有两个换行符
print(line2,end="") # 只有一个换行符
```

```python
# 因此可以使用循环读取整行
with open(file="data/file_learning.txt", encoding="utf-8", mode="r") as f:
    while True:
        line = f.readline()
        if not line :
            break
        print(line, end="")
```

# 写入文件


    
`模式`:

    "w":写入模式，会覆盖原内容
    "a":附加模式，不会覆盖原内容，在文件末尾直接添加
    "r":读取模式
    "r+":以读写方式打开文件。文件的指针将会放在文件的开头。如果文件不存在，抛出异常
    "w+":  以读写方式打开文件。如果文件存在会被覆盖.如果文件不存在,创建新文件
    "a+": 以读写方式打开文件。如果该文件已存在，文件指针将会放在文件的结尾。如果文件不存在，创建新文件进行写入
    
    省略模式实参时，Python将以默认的只读模式打开文件
    要写入的文件不存在，函数 open() 将自动创建它。
    
    

```python
# 使用相对路径："data/file_learning.txt"访问文件
with open(file="data/file_learning.txt", mode="w", encoding='utf-8') as file_object:  # 打开文件
    # 函数 write() 不会在你写入的文本末尾添加换行符，需要手动添加
    file_object.write("pythonw文件学习\n1.1 写入文件") # 写文件
```

```python
file_name = "./data/file_learning.txt"
with open(file=file_name, mode="a", encoding="utf-8") as file_object:
    file_object.write("\n写入文件的模式：\n'w':写入模式，会覆盖原内容\n'a':附加模式，不会覆盖原内容，在文件末尾直接添加")
    
```

```python
def file_copy(file_name1, file_name2):
    """读取file_name1的内容，复制到file_name2中"""
    f1 = open(file_name1, encoding="utf-8", mode="r")
    f2 = open(file_name2, encoding="utf-8", mode="w")
    content = f1.read()
    f2.write(content)
    f1.close()
    f2.close()
```

# 文件指针和seek（）

```python
with open("readme.txt", mode="w+", encoding="utf-8") as f:
    for i in range(5):
        f.write("life is short,i use python")
    # write完后马上读由于指针在末尾，读出为空字符串("")
    content = f.read()
    if content == "":
        print("empty")
```

```python
with open("readme.txt", mode="w+", encoding="utf-8") as f:
    for i in range(5):
        f.write("life is short,i use python")
    # write完后马上读由于指针在末尾，读出为空，故需要使用seek（）
    f.seek(0)# 回到文档首
    content = f.read()
    print(content)
```

```python
with open("readme.txt", mode="w+", encoding="utf-8") as f:
    for i in range(5):
        f.write("life is short,i use python")
    # seek()可以调节指针的位置
    f.seek(3) # 回到文档第3个字符
    content = f.read()
    print(content)
```

```python
with open("data/file_learning.txt", encoding="utf-8", mode="r") as f:
    content = f.read()  # read()后指针已经到达文档尾部
    # 通过seek重新调节位置--注意
    f.seek(7)  
    # 注意：中文+汉字可能出现乱码，第8个是中文"文"
    try:
        c = f.read()
        print(c)
    except Exception as e:
        print(e)
    
```

## 二进制模式

```python
with open("data/file_learning.txt",mode="rb") as fb:  # 以二进制模式打开文件
    content = fb.read()  # 读取的是二进制文件流
    print(type(content))
    print(content)
    print("_"*60)
    fb.seek(8)
    print(fb.read())
```

```python
# 二进制打开文件写入
with open("hello2.txt", mode="wb+") as fb:
    fb.write("你好".encode("utf-8")) #   二进制打开文件写入汉字
    fb.write(b"hello")   #   二进制打开文件写入ASCII码-字符前面加b写入二进制
    # 查看写入内容
    fb.seek(0)
    content = fb.read()
    print(content)
```

## 利用文件管理实现目录的深度优先遍历 

```python
def dfs_dir(path, width):
    """
    :param path: 当前到达的路径
    :parm width: 使用width设置格式化打印输出
    
    """
    # 获取当前路径包含的所有文件
    files = os.listdir(path)
    
    # 依次判断是否为文件夹
    for file in files:
        # 打印文件
        print("-"*width + file)
        next_dir = path+"/"+file
        if os.path.isdir(next_dir):# 如果是文件夹则继续遍历
            dfs_dir(next_dir,width+4 )
            

dfs_dir("D:/yangdaopy/mycode/python/pandas/", 0)
        
```
