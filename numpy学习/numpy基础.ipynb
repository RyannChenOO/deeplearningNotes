# Numpy可以高效处理大数组的数据

这是因为：

NumPy是在一个连续的内存块中存储数据，独立于其他Python内置对象。NumPy的C语言编写的算法库可以操作内存，而不必进行类型检查或其它前期工作。比起Python的内置序列，NumPy数组使用的内存更少。
NumPy可以在整个数组上执行复杂的计算，而不需要Python的for循环。

```python
# nparray计算效率快于python原生数组

# 使用随机数创建一个list-大小为一个亿
my_list = []
for i in range(100000000):
    my_list.append(random.random())
# 求和并输出本次求和所耗费的时间
starttime_mylist = time.time()
sum_mylist = sum(my_list)
endtime_mylist = time.time()

# 使用列表创建一个nparray
my_nparray = np.array(my_list)

# 求和并输出本次求和所耗费的时间
starttime_mynparray = time.time()
sum_mynparray = np.sum(my_nparray)
endtime_mynparray = time.time()

print("原始列表存储时求和耗时{}".format(endtime_mylist-starttime_mylist))
print("numpy数组存储时求和耗时{}".format(endtime_mynparray-starttime_mynparray))
```

基于NumPy的算法要比纯Python快10到100倍（甚至更快），并且使用的内存更少。

# 数组的切片与索引
    实质上是按行操作

一维数组:与list相似

```python
a = np.arange(10)
print(a)
# 索引
print(a[2])
# 切片
print(a[1:7:2]) # start:stop:step
print(a[2:])
```

跟list最重要的区别在于，数组切片是原始数组的视图。这意味着数据不会被复制，视图上的任何修改都会直接反映到源数组上。

```python
arr = np.arange(10)
arr_slice = arr[5:8]
print(arr)
print(arr_slice)
# 修改arr_slice中的值，变动也会体现在原始数组arr中
arr_slice[1] = 12345
print(arr)
print(arr_slice)
```

```python
# 基于原地修改的特性可以给全部数值赋值
arr = np.arange(10)
print(arr[:]) # 所有值
arr_slice = arr[:]
print(arr_slice)
arr_slice[:] = 66
print(arr)
print(arr_slice)
```

如果你想要得到的是ndarray切片的一份<span class="mark">副本而非视图</span>，就需要明确地进行复制操作，例如arr[5:8].copy()

```python
arr = np.arange(10)
arr_slice_copy = arr_slice[5:8].copy()
arr_slice[:] = 55
print(arr)
print(arr_slice_copy)
```

## 数组添加、删除、去重-返回新数组
    添加：append()[末尾添加]，insert()[指定位置添加]
    删除：delete()
    去重：unique

```python
# np.append()-不改变原数组
a = np.array([[1,2,3],[4,5,6]]) # 二维数组（2，3）
print(a)
print("*"*34)
# 直接添加
b = np.append(a, [7,8,9])
print(a)    # a未改变，添加元素是返回到新数组(一维数组)中了
print("{}:{}".format(b.shape,b))
b.shape = (3,3)
print(b)
print("*"*34)

# 按轴添加
c = np.append(a, [[7,8,9]], axis=0) # [[7,8,9]]相当于一个二维数组(1,3)
print(c)
print("*"*34)

d = np.append(a,[[5,5,5,5],[7,8,9,10]], axis=1) # [[5,5,5,5],[7,8,9,10]]相当于二维数组(2,4)
print(d)
print("*"*34)


a = np.array([[1,2],[3,4],[5,6]])
print(a)
print("*"*34)
# 直接添加——新数组展开为一维
b = np.insert(a,2,[7,8])    # 在a20前面添加
print(b)
print("*"*34)
c = np.insert(a,2,[[7,8]], axis=0)
print(c)
print("*"*34)

# 按轴添加
d = np.insert(a, 2, [11], axis=0)   # 添加了[11,11]
print(d)
print("*"*34)

e = np.insert(a, 2, [9,9,9], axis=1)
print(e)
print("*"*34)

f = np.insert(a, 0, [9], axis=1)
print(f)
print("*"*34)

a = np.arange(12).reshape(3,4)
print(a)
print("*"*34)
# 直接删除
b = np.delete(a,1)  # 展开的指定位置进行删除
print(b)
print("*"*34)
# 按轴删除
c = np.delete(a,1, axis=0)
print(c)
print("*"*34)

d = np.delete(a,1, axis=1)
print(d)
print("*"*34)

a = np.array([1,2,3,3,5,2,1,6]).reshape(2,4)
print(a)
print("*"*34)

b = np.unique(a)    # 展开为一维数组
print(b)
print("*"*34)
c, indexes = np.unique(a, return_index=True)    # 去重数组元素在原数组第一次出现时对应的下标
print(c)
print(indexes)
print("*"*34)
d, indices = np.unique(a, return_inverse=True)  # 原数组各元素在新数组出现时对应的下标
print(d)
print(indices)
print("*"*34)

c, counts = np.unique(a, return_counts=True)
print(c)
print(counts)   # 新数组中各元素重复出现的次数
```

## 广播机制

不同大小的数组之间的运算叫做广播（broadcasting）

[broadcasting documentation](https://numpy.org/doc/stable/user/basics.broadcasting.html)

广播机制1：数组与标量的算术运算会将标量值传播到各个元素

```python
A = np.array([[1,2],
              [2,1]])
1 / A, 0.5 * A
```

两个都为矩阵的广播机制

```python

# case 1:减去列平均值的方式对数组的每一列进行距平化处理
arr = np.arange(12).reshape(4,3)
print(arr)  # （4，3）
print("-"*40)
print(arr.mean(axis=0))# （1，3）
print("-"*40)
# （4，3）-- (1,3),低维相同，所以在上一维的维度1(每一列)进行广播
demeaned = arr - arr.mean(axis=0)
print(demeaned)
```

## 通用函数

通用函数（即ufunc）是一种对ndarray中的数据执行元素级运算的函数。

一元（unary）ufunc





```python
# 求平方根
a = np.array([4., 9., 25.])
np.sqrt(a)
b = np.empty_like(a, dtype=np.float64)
# ufunc可以接收一个输出数组
np.sqrt(a, b)
# 实现原地数组操作
np.sqrt(a,a)
a
```

```python
# 返回浮点数数组的小数和整数部分
arr = np.random.randn(7) * 5
print(arr)
remainder, whole_part = np.modf(arr)
print(remainder, whole_part)
```

二元（binary）ufunc:接受2个数组,返回一个结果数组




```python
a = np.array([[11,66],[55,44],[55,33]])
b = np.array([[11,0], [5,22], [88,5]])
# 计算a和b中元素级别最大的元素
print(np.maximum(a,b))

# 可以接收列表,转为ndarray
res = np.maximum([11,22,33,44,55,66],[88,22,33,99,44,77])   # 依次比较取最值
print(type(res))
print(res)
```

网格点

numpy.meshgrid()直接生成想要的网格点:给定X轴和Y轴所有的点，将X和Y放到两个数组，进行笛卡尔乘积，从而得到所有的点。


```python
x = np.array([0, 1, 2])
y = np.array([0, 1])
X, Y = np.meshgrid(x, y)
# X与Y的对应维度组合算一个坐标（0，0），(1,0),(2,0),(0,1),(1,1),(2,1)
print(X)
print(Y)

# 可视化
import matplotlib.pyplot as plt
plt.plot(X,Y, color="r", marker=".", linestyle="")
plt.grid(True)
plt.show()
```

逻辑运算

numpy.where函数是三元表达式x if condition else y的矢量化版本


```python
a = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
b = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
# 使用布尔数组
cond = np.array([True, False, True, True, False])
# 假设我们想要根据cond中的值选取a和b的值：当cond中的值为True时，选取a的值，否则从b中选取
# 如果直接使用列表生成式则速度不是很快（基于纯python）,并且无法处理多维的情况
# 使用np.where
np.where(cond, a, b)
```

np.where的第二个和第三个参数不必是数组，它们都可以是标量值。

```python
# 假设有一个由随机数据组成的矩阵，希望将所有正值替换为2，将所有负值替换为－2
arr = np.random.randn(4,4)
print(arr>0) # 生成布尔数组
np.where(arr > 0, 2, -2)
```

使用np.where，可以将标量和数组结合起来

```python
arr = np.random.randn(4,4)
# 将arr中＞0的同2替换，其他位置保持不变
np.where(arr>0, 2, arr)
```

## 排序



sort方法就地排序,会修改数组本身.

np.sort返回的是数组的已排序副本

```python
a = np.random.randn(2,4)
print(a)
print(a.sort()) # None!!因为sort是就地排序
print(a) # 已排序
```

```python
a = np.random.randn(2,4)
print(a)
print(np.sort(a)) # np.sort()会返回副本
print(a) # 未排序
```

```python
a = np.random.randn(2,4)
print(a)
a.sort(axis=0)
print("-"*40)
print(a) # 已排序
```

## 线性代数

numpy.linalg中有一组标准的矩阵分解运算以及诸如求逆和行列式之类的东西。它们跟MATLAB和R等语言所使用的是相同的行业标准线性代数库，如BLAS、LAPACK、Intel MKL（Math Kernel Library，可能有，取决于你的NumPy版本）等



```python
from numpy.linalg import inv, qr
x = np.arange(10).reshape(2,5)
print("x:\n", x)
print("-"*30)


# 矩阵转置
x_T = x.T
print("x的转置：\n", x_T)


# 矩阵点积（矩阵乘法）
mat = x.T.dot(x) # 计算X和它的转置X.T的点积
print("x的转置x.T 点积 x:\n", mat)
print("-"*30)

# 计算方阵的逆-使用inv--!!注意矩阵必须可逆(A@A(-1)=E)
x = np.random.randn(5,5)
mat = x.T.dot(x)
print("mat = x.T@x,mat的逆：\n", inv(mat))
print("-"*30)
print(mat.dot(inv(mat)))

# 矩阵求行列式|A|
A = np.array([[1,1],
              [0,1]])
np.linalg.det(A)
```

### nan和inf
计算nan个数:arr!=arr与np.count_nonzero(arr!=arr)机制

```python
a = np.nan
b = np.inf
print(a,type(a))
print(b,type(b))

c = np.arange(12, dtype=float).reshape(3,4) # 必须转换为float才能在之后进行nan赋值
c[1:3,:] = a
np.count_nonzero(c) # 计算非False/非0的个数
print(np.nan != np.nan)  # 只有nan比较会是True（1）
print(a != a)
print(b != b) #False（0）

# 利用arr!=arr机制查询、修改nan
print(c != c)   # 数按元素依次判断
print(np.count_nonzero(c!=c))# 计算nan的个数

c[c!=c] = 1 # 修改c中nan为1
c[np.isnan(c)] = 0  # 修改c中nan为0
# 将d中nan全体换成这一类的有效数值的平均值
d = np.arange(24).reshape(4,6).astype("float")
d[1:3,:] = np.nan
print(d)
for i in range(d.shape[1]):  # 按列循环
    col = d[:,i]    # 取该列
    if np.count_nonzero(col != col) > 0:    # 存在nan
        not_nan_col = col[col==col] # 取该列的非nan算均值
        col[col != col] = np.mean(not_nan_col)
```
